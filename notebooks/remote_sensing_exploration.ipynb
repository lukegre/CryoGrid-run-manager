# ASTER for emissivity

```python
from matplotlib import pyplot as plt

from cryogrid_pytools import data
```

```python
bbox = 71.360186,39.475808,71.931818,39.816300
bbox = 71.921997,38.715100,72.597656,39.200982
emissivity = data.get_aster_ged_emmis_elev(bbox)
albedo = data.get_modis_albedo_500m(bbox)
```

```python
fig, axs = plt.subplots(1, 3, figsize=(12, 4), sharex=True, sharey=True)

props = dict(robust=True, cbar_kwargs=dict(location='bottom', pad=0.01, extend='neither'))
emissivity.elevation.plot.imshow(ax=axs[0], cmap='terrain', **props)
emissivity.emissivity.mean('band').plot.imshow(ax=axs[1], cmap='bone_r', **props, vmin=0.92)
albedo.Albedo_BSA_shortwave_p10.plot.imshow(ax=axs[2], cmap='Grays_r', **props, vmin=0)

axs[0].set_title('Elevation (ASTER GED 100 m)')
axs[1].set_title('Emissivity (ASTER GED 100 m)')
axs[2].set_title('Albedo (MODIS 500 m)')

for ax in axs:
    ax.set_xlabel('')
    ax.set_ylabel('')
    ax.set_xticks([])

plt.tight_layout()
```
